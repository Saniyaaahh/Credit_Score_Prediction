# **Applying ML Model**

```python
df['Credit_Mix'].unique()
```

```python
df["Credit_Mix"] = df["Credit_Mix"].map({"Standard": 1, "Good": 2, "Bad": 0})
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = np.array(df[["Annual_Income", "Monthly_Inhand_Salary",
                   "Num_Bank_Accounts", "Num_Credit_Card",
                   "Interest_Rate", "Num_of_Loan",
                   "Delay_from_due_date", "Num_of_Delayed_Payment",
                   "Credit_Mix", "Outstanding_Debt",
                   "Credit_History_Age", "Monthly_Balance"]])
y = np.array(df[["Credit_Score"]])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(xtrain, ytrain)
```

```python
predictions = model.predict(xtest)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
print('Classification Report', classification_report(ytest, predictions))
print('\n')
print('Confusion Matrix', confusion_matrix(ytest, predictions))
print('\n')
print('Accuracy Score', accuracy_score(ytest, predictions))
```

The machine learning model achieved an accuracy score of approximately 80.75%, indicating its overall correctness in predicting credit scores. The classification report provides precision, recall, and F1-score for each class (Good, Poor, Standard), showing the model's performance in distinguishing between different credit score categories. The confusion matrix further visualizes the model's predictions and actual outcomes for each class
